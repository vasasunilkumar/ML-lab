```
import matplotlib.pyplot as plt
import numpy as np
```

```
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```
x = [1,2,3,4,5,6,7,8,9]
y1 = [1,3,5,7,9,11,13,15,20]
y2 = [10,20,30,40,50,60,70,80,90]
plt.plot(x, y1,"y",label="Line l",linewidth=5)
plt.plot(x, y2, "r",label="Line h",linewidth=4)
plt.plot()
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()
plt.show()

```

```
a = np.arange(0,10,0.5)
plt.plot(a,"go:",a**3,"r^--",a**2,"c+-.",label="Lh")
plt.xlabel("X label")
plt.ylabel("Y label")

plt.legend()
plt.show()
```
